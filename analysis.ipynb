```python
import pandas as pd
import sqlite3
from google.colab import files

# Upload CSV
uploaded = files.upload()  # choose Superstore_sale_data.csv

# Load data
df = pd.read_csv("Superstore_sale_data.csv")

# Convert date columns
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True, errors="coerce")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True, errors="coerce")

# Push into SQLite
conn = sqlite3.connect(":memory:")
df.to_sql("superstore", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT
    STRFTIME('%Y', "Order Date") AS year,
    STRFTIME('%m', "Order Date") AS month,
    SUM(Sales) AS total_revenue,
    COUNT(DISTINCT "Order ID") AS order_volume
FROM superstore
GROUP BY year, month
ORDER BY year, month;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT State, SUM(Sales) AS total_revenue
FROM superstore
GROUP BY State
ORDER BY total_revenue DESC
LIMIT 10;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT Category, "Sub-Category", SUM(Sales) AS total_revenue
FROM superstore
GROUP BY Category, "Sub-Category"
ORDER BY total_revenue DESC;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    STRFTIME('%m', "Order Date") AS month,
    SUM(Sales) AS monthly_sales,
    COUNT(DISTINCT "Order ID") AS monthly_orders
FROM superstore
WHERE STRFTIME('%Y', "Order Date") = '2017'
GROUP BY month
ORDER BY month;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    Region,
    SUM(Sales) AS total_revenue,
    COUNT(DISTINCT "Order ID") AS order_volume
FROM superstore
GROUP BY Region
ORDER BY total_revenue DESC;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
    STRFTIME('%Y', "Order Date") AS year,
    SUM(Sales) AS yearly_revenue,
    COUNT(DISTINCT "Order ID") AS yearly_orders
FROM superstore
GROUP BY year
ORDER BY year;
"""
pd.read_sql_query(query, conn)
```
